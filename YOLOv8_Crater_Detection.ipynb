```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("lincolnzh/martianlunar-crater-detection-dataset")

print("Path to dataset files:", path)
```

```python
from ultralytics import YOLO


model = YOLO('yolov8n.pt')


model.train(data='dc.yaml', epochs=50, imgsz=640)
```

```python
# Evaluate the model
results = model.val(data='dc.yaml')
print(results)

```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt


model = YOLO('/content/runs/detect/train3/weights/best.pt')
img_path = '/content/01_png.rf.4d2ebc5ed98ad1e69d667aadbce63d53.jpg'
image = cv2.imread(img_path)

results = model.predict(source=img_path, save=True, conf=0.35)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```
